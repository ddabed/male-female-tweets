```python
import pickle
import pyLDAvis
import pyLDAvis.gensim_models as gensimvis
import gensim
import os
```

```python
path = os.getcwd()+'/Output/'

gensim_dictionary = gensim.corpora.Dictionary.load(path+'gensim_dictionary.gensim')
gensim_corpus = pickle.load(open(path+'gensim_corpus_corpus.pkl', 'rb'))
lda_model = gensim.models.ldamodel.LdaModel.load(path+'gensim_model.gensim')
```

```python
lda_visualization = gensimvis.prepare(lda_model, gensim_corpus, gensim_dictionary, sort_topics=False)
pyLDAvis.display(lda_visualization)
```
